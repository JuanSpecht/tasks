##Importo bibliotecas

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

## Cargo el dataset de Anscombe

```python
url = 'https://gist.githubusercontent.com/JuanSpecht/3f9b120ea6d20d1c51d1d5af8e0975ef/raw/c92c66e43d144fa9c29dbd602d5af6988e8db533/anscombes.csv'
anscombe = pd.read_csv(url)
anscombe
```

```python
# Diccionario para traducir los números romanos que identifican a cada dataset 
romanDict = {
  1: "I",
  2: "II",
  3: "III",
  4: "IV"
}

# Creo los cuatro nuevos puntos
newData = pd.DataFrame(
[[16,8],
[8,6],
[16,10],
[16,6],],
index=["a", "b", "c", "d"],
columns=['x', 'y'])

# Grafico los datasets
fig = plt.figure(figsize=(8, 6), dpi=80)
colors=["blue", "red", "green", "yellow"]

for i in range(1, 5):
    plt.subplot(2, 2, i)
    ds = anscombe[anscombe.dataset == romanDict[i]]
    plt.scatter(ds.x,ds.y, color = "grey")
    plt.xlim(2,20)
    plt.ylim(2,15)
    for ii in range (4):
      plt.scatter(newData.x[ii], newData.y[ii], color = colors[ii])
```

Más allá de que los cuatro datasets tienen parámetros estadísticos idénticos, al graficarlos se puede ver que responden a comportamientos bien diferentes: El primero tiene una distribución aleatoria, el segundo sigue una ley que poidría ser cuadrática, el tercero lineal (con un outlier) y en el último tiene fijo el valor de x (con un outlier)

Al graficar los cuatro nuevos puntos se puede conjeturar a qué dataset es más probable que pertenezca cada uno, teniendo en cuenta que cada uno parece seguir el comportamiento de uno de los datasets de Anscombe:

* El punto <font color='red'> rojo </font> parece pertenercer al cuarto dataset,
* El punto <font color='green'> verde </font>, al tercer dataset,
* El punto <font color='yellow'> amarillo </font>, al segundo,
* Y el punto <font color='blue'> azul </font>, entonces pertenecería al primer dataset.
